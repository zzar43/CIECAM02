```r
require(ggplot2)
```

```r
data <- read.csv("data1.csv", header=T,stringsAsFactors =F)
head(data)
```

```r
ggplot(data, aes(x = environment, y = J)) +
  geom_boxplot() +
  scale_x_discrete() + xlab("Environment") +
  ylab("J")
```

```r
bartlett.test(data$J~data$environment,data=data)
aov.J <- aov(data$J~data$environment, data=data)
summary(aov.J)
```
